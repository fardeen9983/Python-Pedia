## Study Histograms, PDF

```python
# Required modules
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

### Fetch Iris dataset

```python
iris = pd.read_csv("iris.data")

# Separate the three columns

iris_setosa = iris.loc[iris['species'] == "setosa"]
iris_virginica = iris.loc[iris['species'] == "virginica"]
iris_versicolor = iris.loc[iris['species'] == "versicolor"]
```

### Plot a 1-D scatter plot with their petal lengths on same graph on same axis

```python
plt.plot(iris_setosa['petal_length'],np.zeros_like(iris_setosa['petal_length']))
plt.plot(iris_versicolor['petal_length'],np.zeros_like(iris_versicolor['petal_length']))
plt.plot(iris_virginica['petal_length'],np.zeros_like(iris_virginica['petal_length']))

plt.xlabel("Petal Length")
plt.show()
```

### Plotting a histogram (density plot)
Instead of creating a 1-D scatter plot whcih gives very less information
we can simply modify the x-axis expnading the intervals and using the y axis to show the number of points present in the given interval on x axis.

For that we use Seaborn, which will thus generate a Histogram

```python
sns.FacetGrid(iris,hue='species',height=5)\
    .map(sns.distplot,'petal_length')\
    .add_legend()
plt.show()
```

The deep colored smooth lines visible in the histogram are called **PDF (Probability Density Function)**

It is a smoother version of the histogram created using **KDE (Kernel Density Estimation)**

It is very clear from the above histogram that if petal length is less than 2, the species is likely to be Setosa.
But both the histograms and PDFs for the other two species overlap. So only petal length cant be the only parameter defining the species
