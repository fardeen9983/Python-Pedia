matplotlib is a popular plotting library aka used for data visualization
pyplot is the python version for Matlab's plotting features
Like plotting numbers, graphs, points, visualising them,etc

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plot a simple graph


```python
# x axis will be automatically generated like 0,1,2,...
plt.plot([2, 4, 6, 1, 3])

# Place axis labels
plt.xlabel("Indices")
plt.ylabel("Numbers")

# Set graph title
plt.title("Demo")

# Plot the graph
plt.show()
```

### Plot numbers and there squares with grid

```python
a = np.arange(1, 6)
# Plot numbers on x axis and squares on y
plt.plot(a, a ** 2)
# Put labels on the axis
plt.xlabel("Number")
plt.ylabel("Squares")
# Show the grid
plt.grid()
# Plot the graph
plt.show()
```

### Display colored points rather than lines

```python
plt.plot(a, a ** 2, "ro")
# r - red color, 0 - dots instead of curve
plt.grid()
plt.show()
```

### Varied plot visualisation
red squares, blue triangles and green dashes
r,g,b : red,blue,green
s,^,-- : square, triangle, dash

```python
a = np.arange(0., 5., 0.2)
plt.plot(a, a ** 2, "rs", label="^2")
plt.plot(a, a ** 2.25, "b^", label="^2.2")
plt.plot(a, a ** 2.5, "g--", label="^2.5")
# Show legend
plt.legend()
plt.grid()
plt.show()
```

### Specify Linewidth

```python
a = np.arange(1,6)
plt.plot(a,a**2,linewidth=5.0)
plt.show()
```

### Set properties using setp( )

```python
x = np.arange(1,5)
# Fetch the individual lines from plot
lines = plt.plot(x,x**2,x,x*3)

# Set individual line properties
# Python style setting property
plt.setp(lines[0],color='g',linewidth=2.5)
# MATLAB style porperty setup
plt.setp(lines[1],'color','b','linewidth',2.5)
```

### Multiple figures and axes 

```python
# Data
a = np.arange(1,19,2) 
# Create just one figure where we can create multiple plots
plt.figure(1)

# Specify the number of subplots needed
plt.subplot(211)
# 2 - no of rows, 1 - no of columns, 1 - figure part number 
# This commands creates two divisions 2x1 in the figure part 1

# Plot the 1st subplot
plt.grid()
plt.plot(a,np.sin(a),"b-")

# Create subplot for 2nd graph within other half of the figure
plt.subplot(212)
# Plot the 2st subplot
plt.grid()
plt.plot(a,np.cos(a),"r--")

plt.show()
```

### Work with multiple figures 

```python
a = np.arange(1,19,2)

# Create first figure
plt.figure(1)
plt.subplot(211)
plt.plot(a,a**2,"r--")

plt.subplot(212)
plt.plot(a,a**4,"b--")

# Create second figure with no subplots
plt.figure(2)
plt.plot(a,np.sqrt(a),"g-")

# Modify first figure
plt.figure(1)
plt.subplot(211)
plt.title("Powers of a")

plt.show()


```
