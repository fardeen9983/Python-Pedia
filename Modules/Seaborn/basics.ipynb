## Seaborn
It is a plotting library with added features and used together with pyplot

### 2-D color coded scatter plot

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fetch the dataset
iris = pd.read_csv("iris.data")
iris
```

#### Plot the 2-D scatter plkot b/w sepal length and width highlighting different species

FacetGrid accepts and sets parameters on the plot such as :
1. The dataframe to use (iris)
2. The column by which to color the different datapoints (hue)
3. The number of divisions in the x axis (height)
4. plot type (plt.scatter)
5. x axis (sepal_length)
6. y axis (sepal_width)
7. show the legend for different hue classes (add_legend)

```python
# Set plot style
sns.set_style("whitegrid")
# Set plot properties using FacetGrid : Multi-plot grid for plotting conditional relationships
sns.FacetGrid(iris,hue="species",height=4)\
    .map(plt.scatter,"sepal_length","sepal_width")\
    .add_legend()
# Plot the graph
plt.show()
```
