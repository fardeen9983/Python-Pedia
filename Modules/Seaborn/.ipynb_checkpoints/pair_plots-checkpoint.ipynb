## Pair Plots
Pair plots (pairwise scatter plots) are used to represent multi feature plots in 2-D

But they cannot visualise graphs in 3-D and 4-D

```python
# Import required packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Iris dataset
iris = pd.read_csv("iris.data")
```

## Plot pair plots on feature combinations 
4 Features in iris dataset -> 6 possible scatter plots
All 6 scatter plots will be represented by pair plots 

```python
sns.set_style("whitegrid")
# Plot the Pair plot
sns.pairplot(iris,hue="species",height=3)
# The output is a a matrix of plot with axises as the features
# The diagonal plots are the pair plots
plt.show()
```
